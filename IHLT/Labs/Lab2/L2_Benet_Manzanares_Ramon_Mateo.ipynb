```python
import nltk
import pandas as pd
from nltk.metrics import jaccard_distance
from google.colab import drive
from scipy.stats import pearsonr
from string import punctuation

drive.mount('/content/drive')
nltk.download('punkt')
```

```python
dt = pd.read_csv('/content/drive/MyDrive/IHTL_Labs/Lab2/STS.input.SMTeuroparl.txt',sep='\t',header=None)
dt['gs'] = pd.read_csv('/content/drive/MyDrive/IHTL_Labs/Lab2/STS.gs.SMTeuroparl.txt', sep='\t', header=None)
# tokenize all words
punct = list(punctuation)
s0_list = [nltk.word_tokenize(dt[0][s]) for s in range(len(dt[0]))]
s1_list = [nltk.word_tokenize(dt[1][s]) for s in range(len(dt[1]))]

def clear_list(sX_list):
  
  for i in range(len(sX_list)):
    for j in s0_list[i]:
      if j in punct:
        sX_list[i].remove(j)
  return sX_list

```

```python
print(len(s0_list), len(s1_list))
```

```python
gs = [1 -jaccard_distance(set(s0_list[x]), set(s1_list[x])) for x in range(len(s0_list))]
```

```python
dt['jaccard'] = gs
```

```python
dt[:25]
```

```python
pearsonr(dt['gs'], dt['jaccard'])[0]
```

## **Experiment 2 Split strings before tokenize**



```python
dt = pd.read_csv('/content/drive/MyDrive/IHTL_Labs/Lab2/STS.input.SMTeuroparl.txt',sep='\t',header=None)
dt['gs'] = pd.read_csv('/content/drive/MyDrive/IHTL_Labs/Lab2/STS.gs.SMTeuroparl.txt', sep='\t', header=None)
s0_list = [nltk.sent_tokenize(dt[0][s]) for s in range(len(dt[0]))]
s1_list = [nltk.sent_tokenize(dt[1][s]) for s in range(len(dt[1]))]
s0_list = [nltk.word_tokenize(s0_list[s][0]) for s in range(len(s0_list))]
s1_list = [nltk.word_tokenize(s1_list[s][0]) for s in range(len(s1_list))]
gs = [round(1 -jaccard_distance(set(s0_list[x]), set(s1_list[x])), 3) for x in range(len(s0_list))]
```

```python
dt['jaccard'] = gs
pearsonr(dt['gs'], dt['jaccard'])[0]
```
