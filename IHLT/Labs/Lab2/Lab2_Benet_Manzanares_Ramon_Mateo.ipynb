##Installation ans imports

```python
import os
import pandas as pd
import nltk

from argparse import Namespace
from google.colab import drive
from nltk.metrics import jaccard_distance
from scipy.stats import pearsonr
import re

nltk.download('punkt')
```

##Settings

```python
settings = Namespace()

settings.mount_path = "/content/drive"
drive.mount(settings.mount_path, force_remount=True)

settings.project_folder = "Benet_MAI/S1/IHLT/IHLT_Labs/Lab2" #@param ["IHTL_Labs/Lab2", "Benet_MAI/S1/IHLT/IHLT_Labs/Lab2" ]
settings.project_path = os.path.join(os.path.join(settings.mount_path, "MyDrive"), settings.project_folder)
settings.input_filename = "STS.input.SMTeuroparl.txt" #@param {type:"string"}
settings.input_filepath = os.path.join(settings.project_path, settings.input_filename)
settings.gs_filename = "STS.gs.SMTeuroparl.txt" #@param {type:"string"}
settings.gs_filepath = os.path.join(settings.project_path, settings.gs_filename)
```

##Data

```python
#################### Read sentences pairs and gold standard similarities ####################
dt = pd.read_csv(settings.input_filepath,sep='\t', header=None)
dt['gs'] = pd.read_csv(settings.gs_filepath, sep='\t', header=None)
dt.head()
```

##Experiment

```python
#################### Compute jaccard distances ####################
def jaccard_similariy(sentences):
  sent1, sent2 = sentences
  words1 = set(nltk.word_tokenize(sent1))
  words2 = set(nltk.word_tokenize(sent2))
  return 1-jaccard_distance(words1, words2)

dt["jaccard"] = list(map(jaccard_similariy, zip(dt[0], dt[1])))
dt.head()
```

```python
#################### Compute pearson correlation ####################
pearsonr(dt['gs'], dt['jaccard'])[0]
```

## Conclusions
As it is represented by the resulting 0.45 pearson correlation, Jaccard simmilarity with raw words is far from the Gold Standard similarity measure, but shares some correlation.
<br>
It can be deduced that similarity measures based on raw words have many problems. One of the most fragrants is the fact that even nearly identical words (such as "paragraph" and "paragraphs") are considered different.
<br><br>
We expect to learn new and better methods in the future in order to be able to obtain results closer to the Gold Standard.
