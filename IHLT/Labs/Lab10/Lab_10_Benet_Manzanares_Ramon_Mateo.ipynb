## Installations and imports

```python
!pip install spacy==2.1.0
!python -m spacy download en_core_web_sm
!pip install neuralcoref

import spacy
import neuralcoref

nlp = spacy.load("en_core_web_sm")
neuralcoref.add_to_pipe(nlp)
```

## Data

```python
text = "Alice was beginning to get very tired of sitting by her sister on the bank, \
and of having nothing to do: once or twice she had peeped into the book her \
sister was reading, but it had no pictures or conversations in it, ‘and what \
is the use of a book,’ thought Alice ‘without pictures or conversations?’"
doc = nlp(text)
print(text)
```

## Experiments

```python
doc._.coref_clusters
```

## Conclusions
Observing the obtained results, first we notice that the main coreference clusters are found, but not all the references. The most evident is "book", for which the method found its references ("it") but not the main word.<br>
On the other hand, this results does not match with the obtained using the <a href="https://huggingface.co/coref/?text=Alice%20was%20beginning%20to%20get%20very%20tired%20of%20sitting%20by%20her%20sister%20on%20the%20bank%2C%20%20and%20of%20having%20nothing%20to%20do%3A%20once%20or%20twice%20she%20had%20peeped%20into%20the%20book%20her%20%20sister%20was%20reading%2C%20but%20it%20had%20no%20pictures%20or%20conversations%20in%20it%2C%20%E2%80%98and%20what%20%20is%20the%20use%20of%20a%20book%2C%E2%80%99%20thought%20Alice%20%E2%80%98without%20pictures%20or%20conversations%3F%E2%80%99">Hugging Face's version of NeuralCoref</a>, which was able to found additional clusters and references such as "pictures or conversations" (see link).
