##Installations and imports

```python
import nltk
nltk.download('gutenberg')
nltk.download('stopwords')
from nltk.corpus import stopwords
from string import punctuation
```

##Experiment

```python
cp = nltk.corpus.gutenberg.words('blake-poems.txt')
#Take all stopwords
stopwords = stopwords.words('english')
```

```python
#List with all punctuations
punct = list(punctuation)
punct.append('."')
#Extract all stopwords in text
cp_not_stopwords = list(filter(lambda x: x.lower() not in stopwords, cp))
cp_not_stopwords = list(filter(lambda x: x not in punct, cp_not_stopwords))
# Transform all 
cp_not_stopwords = map(lambda x : x.lower(), cp_not_stopwords)
# Create a dict with pair {stopword : occurrences}
non_stopwords_dict = {}
for word in cp_not_stopwords:
  count = non_stopwords_dict.get(word, 0)
  non_stopwords_dict[word] = count + 1
```

```python
# Sort stopwords and take first 25 with more occurrences
non_stopwords_ordered = sorted(non_stopwords_dict.items(), key=lambda x: x[1], reverse=True)
for item in non_stopwords_ordered[:25]:
  print(item)
```
